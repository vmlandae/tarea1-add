```python
import pandas as pd
import numpy as np

df2019=pd.read_csv(r'C:/Users/Javier Vega/Documents/GitHub/tarea1-add/data/interim/d2019.csv' , sep =",")
df2020=pd.read_csv(r'C:/Users/Javier Vega/Documents/GitHub/tarea1-add/data/interim/d2020.txt', sep =" ")
```

```python
df2019=df2019.iloc[:,1:]
df2020=df2020.iloc[:,1:]

df2019
```

```python
month=df2020.columns[1:13]
month
```

```python
df_concat=pd.concat([df2019,df2020])
df_concat
```

```python
df_sii = pd.melt(df_concat,id_vars=["Year",'Día'],value_vars=month,var_name='Mes')
df_sii=df_sii[["Year","Mes","Día","value"]]
df_sii.head()
df_sii.to_csv('C:/Users/Javier Vega/Documents/GitHub/tarea1-add/data/interim/df_sii.csv',
                  sep=';',index=False)
```

```python
df_sii
```

```python
print(df_sii.groupby(by=["Mes"],as_index = True)["value"].agg("mean"))
```

```python
print(df_sii.groupby(by=["Year","Mes"],as_index = False)["value"].agg("mean"))
```

2- Al utilizar as_index= True se toma como columna de agrupación la indicada en el parámetro by, caso contrario se ocupa la primera o index del df.

3- Missing values se infiere corresponde a los fines de semana y feriados donde no se registra valor para el precio del dolar.
