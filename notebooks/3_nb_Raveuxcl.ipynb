```python
import pandas as pd
import numpy as np

df2019=pd.read_csv(r'https://raw.githubusercontent.com/Raveuxcl/tarea1-add/master/data/interim/d2019.csv' , sep =",")
df2020=pd.read_csv(r'https://raw.githubusercontent.com/Raveuxcl/tarea1-add/master/data/interim/d2020.txt', sep =" ")
```

```python
df2019=df2019.iloc[:,1:]
df2020=df2020.iloc[:,1:]

df2019
```

```python
month=df2020.columns[1:13]
month
```

```python
df_concat=pd.concat([df2019,df2020])
df_concat
```

```python
df_concat["Día"] = df_concat["Día"].astype(int)
df_concat["Year"] = df_concat["Year"].astype(int)
df_concat
```

```python
df_sii = pd.melt(df_concat,id_vars=["Year",'Día'],value_vars=month,var_name='Mes')
df_sii=df_sii[["Year","Mes","Día","value"]]
df_sii.head()
df_sii.to_csv('https://raw.githubusercontent.com/Raveuxcl/tarea1-add/master/data/interim/df_sii.csv',
                  sep=';',index=False)
```

```python
def cambio_mes (param):
    if param == "Ene":
        param = "Jan"
    if param == "Feb":
        param = "Feb"
    if param == "Mar":
        param = "Mar"
    if param == "Abr":
        param = "Apr"
    if param == "May":
        param = "May"
    if param == "Jun":
        param = "Jun"
    if param == "Jul":
        param = "Jul"
    if param == "Ago":
        param = "Aug"
    if param == "Sep":
        param = "Sep"
    if param == "Oct":
        param = "Oct"
    if param == "Nov":
        param = "Nov"
    if param == "Dic":
        param = "Dec"
        
    return param
```

```python
df_sii['Mes_en'] = df_sii["Mes"].apply(cambio_mes)
df_sii
```

```python
df_sii['Fecha'] =  (df_sii['Día'].apply(str) +"-"+ df_sii['Mes_en'] +"-"+  df_sii['Year'].apply(str))
df_sii
```

```python
df_sii['Fecha'].values
```

```python
df_sii = df_sii[df_sii['value'].notnull()].loc[:, ['Year','Mes', 'Día', 'value','Fecha']]
df_sii
```

```python
from datetime import datetime

df_sii['Fecha'] = pd.to_datetime(df_sii['Fecha'], format='%d-%b-%Y')
df_sii
```

```python
df_sii['dia_de_la_semana'] = df_sii['Fecha'].dt.day_name() 
df_sii
```

```python
def dia_semana (dia):
    if dia == "Monday":
        dia = "Lunes"
    if dia == "Tuesday":
        dia = "Martes"
    if dia == "Wednesday":
        dia = "Miércoles"
    if dia == "Thursday":
        dia = "Jueves"
    if dia == "Friday":
        dia = "Viernes"
    if dia == "Saturday":
        dia = "Sábado"
    if dia == "Sunday":
        dia = "Domingo"
        
    return dia
```

```python
df_sii['dia_de_la_semana'] = df_sii["dia_de_la_semana"].apply(dia_semana)
df_sii
```
