**Importar base de datos 'dolar2019_sii.csv'**: para que se importe bien, debe especificarse el separador, que en este caso es ';', lo cual se hace con sep=';', ya que el archivo csv tiene las columnas separadas por ;.

```python
dolar_2019 = pd.read_csv('https://raw.githubusercontent.com/CatitaR/tarea1-add/master/data/raw/dolar2019_sii.csv',
                         sep=';')
dolar_2019
```

```python
#creo una lista 'names', que me servirÃ¡ despuÃ©s para darle los mismos nombres a todos los data set
names = dolar_2019.columns
```

```python
#filtro sacando la Ãºltima fila
dolar_2019 = dolar_2019[dolar_2019['DÃ­a'] != 'Promedio']
#convierto la primera fila de la variable a tipo 'int' y las siguientes a tipo 'float', reemplazando la coma por punto
for i in dolar_2019:
    if i=='DÃ­a':
        dolar_2019[i] = dolar_2019[i].str.replace(',', '.').astype(int)
    else:
        dolar_2019[i] = dolar_2019[i].str.replace(',', '.').astype(float)
#creo la variable 'year', que toma el valor 2019 y es de tipo int64
dolar_2019['year'] = 2019
dolar_2019['year'] = dolar_2019['year'].astype('int64')
dolar_2019
```

**Importar la base de datos 'dolar_2020_sii.txt'**: tambiÃ©n es un archivo de texto plano, sin embargo, a diferencia de 'dolar_2019_sii.csv', estÃ¡ separado por la lÃ­nea vertical (|) en lugar de punto y coma (;). AdemÃ¡s, en el archivo 'dolar_2020_sii.txt' la primera fila son datos, no los nombres de las filas. Por lo tanto, para que la funciÃ³n read_csv lea bien el archivo, se debe hacer lo siguiente:
- Especificar que el separador es la lÃ­nea vertical, lo cual se hace con sep='|'
- Indicar que la primera fila no corresponde al nombre de las columnas, con header=None
- Indicar cuÃ¡les serÃ¡n los nombres de las columnas, para que la base de datos se pueda entender mejor. Esto se hace con el parÃ¡metro 'names', y en este caso le indicamos que los nombres de las columnas sean los mismos de la base de datos dolar_2019

```python
dolar_2020_txt = pd.read_csv('https://raw.githubusercontent.com/CatitaR/tarea1-add/master/data/raw/dolar_2020_sii.txt',
                            sep='|',
                            header=None,
                            names=names)
dolar_2020_txt
```

```python
#filtro sacando la Ãºltima fila
dolar_2020_txt = dolar_2020_txt[dolar_2020_txt['DÃ­a'] != 'Promedio']
#convierto la primera fila de la variable a tipo 'int' y las siguientes a tipo 'float', reemplazando la coma por punto
for i in dolar_2020_txt:
    if i=='DÃ­a':
        dolar_2020_txt[i] = dolar_2020_txt[i].str.replace(',', '.').astype(int)
    else:
        dolar_2020_txt[i] = dolar_2020_txt[i].str.replace(',', '.').astype(float)
#creo la variable 'year', que toma el valor 2020 y es de tipo int64
dolar_2020_txt['year'] = 2020
dolar_2020_txt['year'] = dolar_2020_txt['year'].astype('int64')
dolar_2020_txt
```

El archivo 'dolar_2020_sii_AA.dat' es de texto plano, pero tiene letras para separar las columnas. Los separadores son distintos segÃºn cada fila, pero tienen la siguiente estructura: una "X", seguido por otra letra del abecedario (distinta a la "X", en mayÃºscula o minÃºscula), seguido por otra "X". Por ejemplo, el separador de la primera fila de "XAX". 

AdemÃ¡s, la base de datos no tiene nombres de las columnas. Por lo tanto, para importarla se debe especificar header=None.

Por otro lado, los decimales estÃ¡n separados por ",", por lo que para importarlo hay que usar sep=';', de lo contrario considera ',' como separador, y eso da error.

Al importar la base de datos, queda con una sola columna con toda la informaciÃ³n. 


```python
dolar_2020_regex = pd.read_csv('https://raw.githubusercontent.com/CatitaR/tarea1-add/master/data/raw/dolar_2020_sii_AA.dat',
                              sep=';',
                              header=None)
dolar_2020_regex
```

```python
#aplico una expresiÃ³n regular, que reemplaza el patrÃ³n X+letra+X por ";". 
#La letra puede estar en mayÃºscula o minÃºscula, puede ser una Ã± o puede tener tildes
dolar_2020_regex[0].replace(r'X[a-zA-ZÃ-Ã¿\u00f1\u00d1]X',';',regex=True, inplace = True)
dolar_2020_regex
```

```python
#Ahora hago un split de la base de datos, separando la columna 0 con el caracter ';'.
#Cambio los nombres de las columnas para que sean los mismos de la base 'dolar_2019'
dolar_2020_regex = pd.DataFrame(dolar_2020_regex[0].str.split(';', expand=True))
dolar_2020_regex.columns = names
dolar_2020_regex
```

```python
#Reemplazo las celdas en blanco con NaN, de lo contrario no me deja convertirlo a float
dolar_2020_regex = dolar_2020_regex.apply(lambda x: x.str.strip()).replace('', np.nan)
#filtro sacando la Ãºltima fila
dolar_2020_regex = dolar_2020_regex[dolar_2020_regex['DÃ­a'] != 'Promedio']
#convierto la primera fila de la variable a tipo 'int' y las siguientes a tipo 'float', reemplazando la coma por punto
for i in dolar_2020_regex:
    if i=='DÃ­a':
        dolar_2020_regex[i] = dolar_2020_regex[i].str.replace(',', '.').astype(int)
    else:
        dolar_2020_regex[i] = dolar_2020_regex[i].str.replace(',', '.').astype(float)
#creo la variable 'year', que toma el valor 2020 y es de tipo int64
dolar_2020_regex['year'] = 2020
dolar_2020_regex['year'] = dolar_2020_regex['year'].astype('int64')
dolar_2020_regex
```
